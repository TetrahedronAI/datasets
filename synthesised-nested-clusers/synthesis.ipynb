```python
FEATURES = 2
CLUSTERS = 15
STANDARD_DEVIATION = 1.5
N_IN_CLUSTER = 1000
```

```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.style import use

use("ggplot")
```

```python
cores = np.array([[random.randint(1, 100) for _ in range(FEATURES)] for _ in range(CLUSTERS)])

plt.scatter(cores.T[0], cores.T[1])
plt.show()
```

```python
data = []
y = []
for label, i in enumerate(cores):
	for _ in range(N_IN_CLUSTER):
		data.append([])
		y.append(label)
		for j in i:
			data[-1].append(round(random.normalvariate(j, STANDARD_DEVIATION), 3))

data = np.array(data)

print("Number of Data Points: ", len(data))

plt.figure(figsize=(24, 10))
plt.scatter(x=data.T[0], y=data.T[1], c=y, cmap="summer")
plt.show()
```

```python
csv = pd.DataFrame(zip(y, *data.T.tolist()))
csv.index.name = "id"
csv.to_csv("synthesised.csv")
csv
```
